```python
# Importing required libraries
import pandas as pd
import numpy as np
import re
import string
import os

# Display settings for better readability in notebooks
pd.set_option('display.max_colwidth', 150)
```

```python
# Define file paths
train_file_path = 'PUBHEALTH/train.tsv'
dev_file_path = 'PUBHEALTH/dev.tsv'
test_file_path = 'PUBHEALTH/test.tsv'

# Load the datasets
train_df = pd.read_csv(train_file_path, sep='\t')
dev_df = pd.read_csv(dev_file_path, sep='\t')
test_df = pd.read_csv(test_file_path, sep='\t')

# Display the first few rows of each dataset to understand the structure
print("Train Dataset:")
display(train_df.head())

print("\nDev Dataset:")
display(dev_df.head())

print("\nTest Dataset:")
display(test_df.head())
```

```python
# Display column names for each dataset
print("Train Dataset Columns:", train_df.columns)
print("Dev Dataset Columns:", dev_df.columns)
print("Test Dataset Columns:", test_df.columns)
```

```python
# Define data cleaning function
def clean_text(text):
    """
    Function to clean text by:
    - Converting to lowercase
    - Removing punctuation
    - Removing special characters
    - Tokenizing words
    """
    if isinstance(text, str):  # Ensure the input is a string
        text = text.lower()  # Convert to lowercase
        text = re.sub(f"[{string.punctuation}]", "", text)  # Remove punctuation
        text = re.sub(r'\d+', '', text)  # Remove numbers
        text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
        return text.strip()
    else:
        return text  # Return as-is if not a string

# Test the cleaning function on a sample text
sample_text = "COVID-19 Vaccine is 95% effective! #Health #Pandemic"
cleaned_text = clean_text(sample_text)
print("Original Text:", sample_text)
print("Cleaned Text:", cleaned_text)
```

```python
# Apply the cleaning function to relevant text columns
train_df['claim_cleaned'] = train_df['claim'].apply(clean_text)
train_df['explanation_cleaned'] = train_df['explanation'].apply(clean_text)
train_df['main_text_cleaned'] = train_df['main_text'].apply(clean_text)

dev_df['claim_cleaned'] = dev_df['claim'].apply(clean_text)
dev_df['explanation_cleaned'] = dev_df['explanation'].apply(clean_text)
dev_df['main_text_cleaned'] = dev_df['main_text'].apply(clean_text)

test_df['claim_cleaned'] = test_df['claim'].apply(clean_text)
test_df['explanation_cleaned'] = test_df['explanation'].apply(clean_text)
test_df['main_text_cleaned'] = test_df['main_text'].apply(clean_text)

# Display cleaned text samples
print("Cleaned Train Dataset:")
display(train_df[['claim', 'claim_cleaned', 'explanation', 'explanation_cleaned']].head())
```

```python
# Check for missing values
print("Missing Values in Train Dataset:")
print(train_df.isnull().sum())

print("\nMissing Values in Dev Dataset:")
print(dev_df.isnull().sum())

print("\nMissing Values in Test Dataset:")
print(test_df.isnull().sum())

# Drop rows with missing values in relevant columns
train_df.dropna(subset=['claim_cleaned', 'explanation_cleaned', 'main_text_cleaned'], inplace=True)
dev_df.dropna(subset=['claim_cleaned', 'explanation_cleaned', 'main_text_cleaned'], inplace=True)
test_df.dropna(subset=['claim_cleaned', 'explanation_cleaned', 'main_text_cleaned'], inplace=True)

# Verify that missing values are handled
print("\nAfter Dropping Missing Values:")
print("Train:", train_df.shape)
print("Dev:", dev_df.shape)
print("Test:", test_df.shape)
```

```python
# Create directory for processed data if it doesn't exist
os.makedirs('data/processed', exist_ok=True)

# Save the cleaned data as new CSV files for modeling
train_df.to_csv('data/processed/train_cleaned.csv', index=False)
dev_df.to_csv('data/processed/dev_cleaned.csv', index=False)
test_df.to_csv('data/processed/test_cleaned.csv', index=False)

print("Cleaned datasets saved successfully!")
```
